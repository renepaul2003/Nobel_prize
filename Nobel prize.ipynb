# Nobel  Prize Winners

### A brief history of the Nobel Prize

When the Swedish businessman Alfred Nobel passed away in 1896, he left behind what was then one of the world’s largest private fortunes. In his last will, Nobel declared that the whole of his remaining fortune of 31.5 million Swedish crowns was to be invested in safe securities and should constitute a fund "the interest on which shall be annually distributed in the form of prizes to those who, during the preceding year, shall have conferred the greatest benefit on mankind".

He will specified in which fields the prizes should be awarded – physics, chemistry, medicine or physiology, literature and peace – and which criteria the respective prize committees should apply when choosing their prize recipients. According to the will, the Nobel Peace Prize was to be awarded “to the person who shall have done the most or the best work for fraternity between the nations and the abolition or reduction of standing armies and the formation and spreading of peace congresses.”

How are the distribution of the ages?.. What is the age of most recipients of a Nobel Prize?

```python
# plotting the graph by histgrom
plt.hist(data['age_get_prize']);
plt.xlim(0, 100);
plt.xlabel('age when they get Nobel Prize');
plt.ylabel('awarded people Count')
plt.title("ages distribution", fontsize=22);
```

Observation:It shows that most people who win a Nobel Prize are between their 60s and 70s

What is the percentage of women and men who won the Nobel Prize?

```python
#plotting the graph by pie graph 
plt.pie(data['gender'].value_counts(), labels=data['gender'].value_counts().index,counterclock=False,autopct='%.2f')
plt.title('awarded people BY GENDER');
```

Observation: It appears that about six percent of Nobel Prize winners are women. I expected that the percentage of women would be less than men, but I did not expect that the difference would be this big.

What is the field with the largest number of Nobel Prize winners?

```python
base_color = sns.color_palette()[0]
order= data['category'].value_counts().index
sns.countplot(data=data,y='category',order=order, color=base_color)
plt.title('awarded people BY category',size=15);
```

Observation:The Nobel Prize is given more to people in the field of medicine

What is the country with the largest number of Nobel Prize winners?

```python
plt.figure(figsize=[20,30])
order= data['born_country_code'].value_counts().index
sns.countplot(data=data,y='born_country_code',order=order, palette = 'CMRmap')
plt.title('awarded people BY COUNTRY',size=20);
```

Observation:The United States has the Most of the Nobel Prizes

What are the ages of Nobel Prize winners in all fields, especially women?

```python
plt.figure(figsize=(20, 5))
sns.swarmplot(x='category', y='age_get_prize',hue = 'gender', dodge=True , data=data, palette = 'CMRmap')
plt.ylabel('Age when get prize')
plt.xlabel('fields')
plt.title('Every winner age seperated by gender and prize category')
plt.show()
```

Observation:This shows that a very small percentage of women won the Nobel Prize in (physics,chemistry, economics). Many young ladies have won the Nobel Prize before the age of thirty-five.
